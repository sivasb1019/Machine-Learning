Malplotlib:

*   Useful for making plots and graphs



```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy library to get datas for our plots
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# sine wave
plt.plot(x, y)
# plt.show()
```

```python
#cosine wave
plt.plot(x, z)
```

```python
# adding title, x-axis & y-axis label
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
```

```python
# plotting parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
```

```python
# plottin the graph with red color and '+' symbol
plt.plot(x, y, 'r+')
```

```python
# plottin the graph with green color and '.' symbol
plt.plot(x, y, 'g.')
```

```python
# plottin the graph with yello color and 'x'
plt.plot(x, y, 'yx')
```

```python
# plotting sin as well as cos wave in a same plot
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Hindi','Tamil', 'English',  'Japanese', 'Spanish']
people = [70, 500, 300, 120, 40]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['Hindi','Tamil', 'English',  'Japanese', 'Spanish']
people = [90, 270, 180, 120, 40]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot - useful in clustering application

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z, color='b')
plt.show()
```

3D Scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z, cmap='Blues')
plt.show()
```
