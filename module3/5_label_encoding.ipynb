Label Encoding:


*   Converting the labels into numeric form.
*    When we are working in a classification problem, we change the label into nemeric.




```python
# importing libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# loading the csv data into pandas dataframe
cancer_data = pd.read_csv("/content/data.csv")
```

```python
cancer_data.head()
```

```python
cancer_data.shape
```

```python
# Finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
# loading the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
print(labels)
```

```python
# appending the labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 --> Benign

1 --> Malignant

```python
cancer_data['target'].value_counts()
```
