Data Standardization:

The process of standardizing the data to a  common format and common range.

```python
# importing libraries
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler # to standardize entire dataset
from sklearn.model_selection import train_test_split # helps to split our data into training and testing data
```

```python
# loading datasets
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
# loading the data into a pandas dataframe
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df.head()
```

```python
df.shape
```

```python
X = df
Y = dataset.target
```

```python
print(X)
```

```python
print(Y)
```

Spliting the data into training data and test data before standardization.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

Standardize the data

```python
# checking whether all the data is in same range. If we get 1, it means we have same range
print(dataset.data.std())
```

```python
# z= (x-u)/s, u - mean, s - standard deviation
scaler = StandardScaler()
```

```python
scaler.fit(X_train) # StandardScaler will understand, how the data distributed in X_train
```

```python
X_train_standardized = scaler.transform(X_train)
```

```python
print(X_train_standardized)
```

```python
X_test_standardized = scaler.transform(X_test)
```

```python
print(X_test_standardized)
```

```python
print(X_train_standardized.std())
```

```python
print(X_test_standardized.std()) # close to 1, which is better than 228
```
