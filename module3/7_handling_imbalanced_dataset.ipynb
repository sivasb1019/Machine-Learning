Imbalanced Dataset:

A dataset with an unequal class distribution.

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# load the csv data into pandas dataframes
credit_card_data = pd.read_csv("/content/credit_data.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.isnull().sum()
```

```python
# determine the distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset.

0 --> Legit Transactions

1 --> Fraudulent Transactions

```python
# seperating the legit and fraud transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Implementing a technique 'Under-Sampling'.

Building a sample dataset containing similar distribution of legit and fraudulent transactions.

Number of Fraudulent Transactions - 492

Number of Legit Transactions - 284315

```python
# creating a ne df with 492 row of legit transations
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

concatenate the two dataframes - legit_sample, fraud

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.shape
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

Now we have evenly distributed datasets for two classes. When we use this dataset for our prediction this will give a better result.
